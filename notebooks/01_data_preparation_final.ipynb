# Notebook 01 - Final Financial ML Data Preparation Pipeline

## Objective

Prepare Iranian stock market data for a leakage-controlled machine learning workflow.

This notebook:

1. Reads raw stock CSV files.
2. Validates required columns.
3. Separates:
   - metadata columns (`insCode`, `dEven`)
   - model features
   - target (`class`)
   - evaluation-only variables
4. Creates clean datasets for the next temporal split stage.

## Label definition

Binary classification:

- `class = 1`
    - Upper barrier (+15%) reached
    - OR positive return without barrier hit

- `class = 0`
    - Lower barrier (-15%) reached
    - OR negative return without barrier hit
    - OR unavailable future information observations

## Important leakage control

The following variables are NOT model features:

- max_price_find
- min_price_find
- after_days
- days_max_find

They are preserved only for future evaluation and censoring analysis.

`dEven` is preserved only for chronological splitting in Notebook 02.


```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
ROOT = Path(r"E:\Iran_stock_trade")

RAW_DATA = ROOT / "raw_data"

OUTPUT_MODEL = ROOT / "data_ready" / "prepared"

OUTPUT_EVAL = ROOT / "data_ready" / "evaluation"


OUTPUT_MODEL.mkdir(parents=True, exist_ok=True)
OUTPUT_EVAL.mkdir(parents=True, exist_ok=True)


print("Raw data:", RAW_DATA)
print("Model output:", OUTPUT_MODEL)
print("Evaluation output:", OUTPUT_EVAL)
```

```python
META_COLUMNS = [
    "insCode",
    "dEven"
]


FEATURE_COLUMNS = [

    "adj_open",
    "adj_high",
    "adj_low",
    "adj_last_price",

    "zigzag_up_new_2",
    "zigzag_down_new_2",

    "EMA_3",
    "EMA_5",
    "EMA_8",
    "EMA_10",
    "EMA_12",
    "EMA_15",
    "EMA_30",
    "EMA_35",
    "EMA_40",
    "EMA_45",
    "EMA_50",
    "EMA_60",
    "EMA_200",

    "RSI_14",

    "macd",

    "power_of_buy",
    "volume_per_avg_month",

    "ho_buy",
    "ho_sell",

    "x",
    "y",
    "color",

    "gmma",
    "started",
    "body"
]


TARGET_COLUMN = [
    "class"
]


EVALUATION_COLUMNS = [
    "max_price_find",
    "min_price_find",
    "after_days",
    "days_max_find"
]


ALL_REQUIRED = (
    META_COLUMNS
    + FEATURE_COLUMNS
    + TARGET_COLUMN
    + EVALUATION_COLUMNS
)


print("Number of model features:", len(FEATURE_COLUMNS))
```

```python
csv_files = list(RAW_DATA.glob("*.csv"))

print("Number of CSV files:", len(csv_files))
```

```python
sample_file = csv_files[0]

sample_df = pd.read_csv(sample_file)

print(sample_file.name)
print(sample_df.shape)

sample_df.head()
```

```python
def prepare_stock_file(file_path):

    df = pd.read_csv(file_path)


    missing_columns = [
        c for c in ALL_REQUIRED
        if c not in df.columns
    ]


    if missing_columns:
        return None, None, missing_columns


    model_df = df[
        META_COLUMNS
        + FEATURE_COLUMNS
        + TARGET_COLUMN
    ].copy()


    evaluation_df = df[
        META_COLUMNS
        + EVALUATION_COLUMNS
    ].copy()


    return model_df, evaluation_df, None
```

```python
success = 0
errors = []


for file in csv_files:

    model_df, evaluation_df, error = prepare_stock_file(file)


    if error:

        errors.append({
            "file": file.name,
            "missing_columns": error
        })

        continue


    name = file.stem


    model_df.to_csv(
        OUTPUT_MODEL / f"{name}_model.csv",
        index=False,
        encoding="utf-8-sig"
    )


    evaluation_df.to_csv(
        OUTPUT_EVAL / f"{name}_eval.csv",
        index=False,
        encoding="utf-8-sig"
    )


    success += 1


print("Completed:", success)
print("Errors:", len(errors))
```

```python
error_report = pd.DataFrame(errors)

error_report
```

```python
prepared_files = list(
    OUTPUT_MODEL.glob("*_model.csv")
)

evaluation_files = list(
    OUTPUT_EVAL.glob("*_eval.csv")
)


print("Prepared model files:", len(prepared_files))
print("Evaluation files:", len(evaluation_files))
```
